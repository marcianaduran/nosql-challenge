```python
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene':{'$eq':20}}

# Use count_documents to display the number of documents in the result
print('Number of establishments with a hygiene score equal to 20: ', establishments.count_documents(query))

# Display the first document in the results using pprint
results = establishments.find(query)
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_eq_20 = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(hygiene_eq_20))

# Display the first 10 rows of the DataFrame
hygiene_eq_20.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'LocalAuthorityName':{'$regex':'London'},'scores.Hygiene':{'$gte':4}}
results = establishments.find(query)

# Use count_documents to display the number of documents in the result
print('Number of establishments in London with a hygiene score equal to 20: ', establishments.count_documents(query))

# Display the first document in the results using pprint
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
rating_gte_4 = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(rating_gte_4))

# Display the first 10 rows of the DataFrame
rating_gte_4.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
ref_latitude = establishments.find_one({'BusinessName':'Penang Flavours'},{'geocode.latitude':1,'_id':0})
ref_longitude = establishments.find_one({'BusinessName':'Penang Flavours'},{'geocode.longitude':1,'_id':0})
ref_longitude
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = ref_latitude['geocode']['latitude']
longitude = ref_longitude['geocode']['longitude']

query = {'RatingValue':{'$eq':5},
         'geocode.latitude':{'$lte':latitude+degree_search},
         'geocode.longitude':{'$lte':longitude+degree_search}}
sort = [('scores.Hygiene', -1)]
limit = 5

# Cast the results as a list and save the results to a variable
results = list(establishments.find(query).sort(sort).limit(limit))

#print results
pprint(results)
```

```python
# Convert result to Pandas DataFrame
df = pd.DataFrame(results)
df.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

match_query = {'$match':{'scores.Hygiene':{'$eq':0}}}
# Write an aggregation query that counts the number of documents and 
# grouped by "LocalAuthorityName"
group_query = {'$group':{'_id':'$LocalAuthorityName',
                        'count':{'$sum':1}}}
# Create a dictionary that will allow the pipeline to sort by count in descending order
sort_values = {'$sort':{'count':-1}}

# Put the pipeline together
pipeline = [match_query, group_query, sort_values]

# Print the number of documents in the result
results = list(establishments.aggregate(pipeline))
print('Number of establishments in each Local Authority area in results', len(results))

# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print(len(df))
# Display the first 10 rows of the DataFrame
df[0:10]
```
